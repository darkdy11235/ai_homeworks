# Bai 1: Tạo 10 điểm dữ liệu ngẫu nhiên và vẽ chúng lên trục toạ độ.

```python
import numpy as np
from matplotlib import pyplot as plt_ex1
from matplotlib import pyplot as plt_ex4
from matplotlib import pyplot as plt_ex7
from sklearn.linear_model import LinearRegression

# exercise 1: generate 10 random points and plot them on the coordinate axes
# generate 10 random points
x = np.random.rand(10)
y = np.random.rand(10)
# plot the points on the coordinate axes
plt_ex1.scatter(x, y)
plt_ex1.title("exercise 1")
plt_ex1.show()
```

# Bai 2: Tính w0, w1 và R-squared dùng thư viện Sklearn

```python
# exercise 2: calculate intercept, slope and R-squared by Scikit-learn library
# reshape the arrays to 2D arrays
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
# create a linear regression model
model = LinearRegression()
# fit the model to the data
model.fit(x, y)
print("\nexercise 2:")
print("w0 = ", model.intercept_)
print("w1 = ", model.coef_)
print("R-squared = ", model.score(x, y))
```

# Bai 3: Tính w0 và w1 với công thức ở slide số 9 và in ra.

```python
# exercise 3: calculate intercept, slope and R-squared by recipe
# calculate intercept and slope
n = len(x)
w1 = (np.sum(x*y)-np.sum(x)*np.sum(y)/n)/(np.sum(x*x)-np.square(np.sum(x))/n)
w0 = np.sum(y)/n - w1*np.sum(x)/n
print("\nexercise 3:")
print("w0 = ", w0)
print("w1 = ", w1)
```

# Bai 4: Vẽ đường thẳng với w0 và w1

```python
# exercise 4: draw the line with the given intercept and slope
y_pred = w0 + w1*x
plt_ex4.scatter(x, y)
plt_ex4.plot(x, y_pred, color='red')
plt_ex4.title("exercise 4")
plt_ex4.show()
```

# Bai 5: Tính R-squared với công thức ở slide số 10 và in ra.

```python
# exercise 5: calculate R-squared by recipe
# R_squared = 1 - Residual_sum_of_squares/Total_sum_of_squares
R_squared = 1 - np.sum(np.square(y - y_pred))/np.sum(np.square(y - np.mean(y)))
print("\nexercise 5:")
print("R-squared = ", R_squared)
```

# Bai 6: Tạo thêm 1 điểm dữ liệu cách rất xa 10 điểm dữ liệu ban đầu và tính lại w0 và w1(công thức) và in ra.

```python
# exercise 6: create a poin very far from the points and recalculate the intercept and slope by recipe
# generate values x of the new point
new_x = np.random.rand(1)
# generate values y of the new point by recipe
# this line is the line perpendicular to the line of the points
new_y = w0 -(1/w1)*new_x+10 # add 10 to make the point very far from the points
x = np.append(x, new_x)
y = np.append(y, new_y)
n = len(x)
w1 = (np.sum(x*y)-np.sum(x)*np.sum(y)/n)/(np.sum(x*x)-np.square(np.sum(x))/n)
w0 = np.sum(y)/n - w1*np.sum(x)/n
print("\nexercise 6:")
print("w0 = ", w0)
print("w1 = ", w1)
```

# Bai 7: Vẽ đường thẳng mới với w0 và w1 mới.

```python
# exercise 7: draw the line with the new intercept and slope
y_pred = w0 + w1*x
plt_ex7.scatter(x, y)
plt_ex7.plot(x, y_pred, color='red')
# set the new point to red
plt_ex7.scatter(new_x, new_y, color='red')
plt_ex7.title("exercise 7")
plt_ex7.show()
```
